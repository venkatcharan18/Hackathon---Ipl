## 10. To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?
## ans - Line plot

```python
de = da.groupby('batsman')['batsman_runs'].sum()
dee = pd.DataFrame(de)
dee
```

```python
tot_runs = dee.sort_values('batsman_runs',ascending = False)
top_10 = tot_runs.head(10)
top_10
```

```python
plt.figure(figsize=(12,10))
plt.xticks(rotation=90)
sns.lineplot(data=top_10,x='batsman',y='batsman_runs')
```

## 11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?
## ans-CHRIS GAYLE

```python
runs = dt.loc[(dt['batsman_runs']==4) | (dt['batsman_runs']==6)]
runs
```

```python
tot = runs.groupby('batsman')['batsman_runs'].sum()
tot
```

```python
tot_runs = pd.DataFrame(tot)
tot_runs.sort_values('batsman_runs',ascending=False).head(3)
```
